### Data Preprocessing

#### 1. One-Hot Encoding
- The sepcies column has categorical features.
- Furthermore, the species in themselves don't seem to have any ranks among themselves (I assume).
- Therefore, we'll one-hot encode the species column.

```python
# One-hot encode the species feature.
df_encoded = pd.get_dummies(df)
df_encoded

# Replace the boolean values in the dataframe with 1s and 0s. 
df_encoded.replace({True : 1, False : 0}, inplace=True)
df_encoded
```

#### 2. Shuffling the training exmaples
- As we're applying a supervised learning model, it requires the dataset to not contain any sorts of ordering in itself.
- Becaue in case if the dataset does contain any implicit ordering, then the model won't be able to generalise well.

```python
# Shuffling to undo any kind of ordering in the dataset.
df_encoded = df_encoded.sample(frac=1)
```

```python
# Separating features and target values 
X, y = df_encoded.iloc[:, 1:], df_encoded["Weight"]

# Printing the shapes of X and y
print(f"Shape of X : {X.shape}")
print(f"Shape of y : {y.shape}")
```

### Feature Engineering
As we're trying to predict the weight of the fish, some of the features that might be more indicative of it are :
- Average length of the fish = (length1 + length2 + length3)/3.
- Approximate volume of the fish = Avg.length\*width\*height.
- The lengths, height, width, all of them seem to have a quadratic relation with the weights, therefore, we can add features by raising them to the power 2.

```python
# Creating new features which are more indicative of the weights of a fish.

# Adding volume and average length to the features.
X["Average_length"] = (X["Length1"]+X["Length2"] + X["Length3"])/3
X["Approximate_volume"] = X["Average_length"]*X["Width"]*X["Height"]

# Adding Squared features
X["Average_length^2"] = X["Average_length"]**2
X["Average_length^4"] = X["Average_length"]**4
X["Height^2"] = X["Height"]**2
X["Width^2"] = X["Width"]**2
```

### K-fold Crossvalidation

The reason behind applying k-folds is that the dataset is pretty small and any other splits of the dataset like 90-10, 80-20 etc..
would leave out very less examples either for testing or for training. But with k-fold crossvalidation we'll be utilizing the entire dataset for both purposes-training and validation.

```python
# k value for the dividing the dataset into k equal parts.
k = 4

# Convert the features and target values to numpy arrays.
X_, y_ = np.array(X), np.array(y)

# Batch size and number of training examples in the dataset.
batch_size = X.shape[0]//k
n = X.shape[0]

# Lists to store errors and R2 scores in all the iterations.
error_train = []
error_test = []
r2_train = []
r2_test = []

for i in range(0, X.shape[0], X.shape[0]//k) :
    train_indices = [k for k in range(0, i)] + [k for k in range(i+batch_size, n)]
    X_test, y_test = X_[i:i+batch_size, :], y_[i:i+batch_size]
    X_train, y_train = X_[train_indices, :], y_[train_indices]
    
    # Standardize the training set and test set.
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train the model.
    model = LinearRegression()
    model.fit(X_train_scaled, y_train)

    # Make predictions and calculate the error.
    pred_train = model.predict(X_train_scaled)
    pred_test = model.predict(X_test_scaled)

    err_train = np.sqrt(mean_squared_error(y_train, pred_train))
    err_test = np.sqrt(mean_squared_error(y_test, pred_test))

    r2_train.append(r2_score(y_train, pred_train))
    r2_test.append(r2_score(y_test, pred_test))
    
    error_train.append(err_train)
    error_test.append(err_test)
```

```python
# Mean training error and testing error.
print(f"Average training error : {sum(error_train)/len(error_train)}")
print(f"Average testing error : {sum(error_test)/len(error_test)}")

# Average R2 scores of training and tesing sets.
print(f"Average training r2 scores : {sum(r2_train)/len(r2_train)}")
print(f"Average testing r2 scores : {sum(r2_test)/len(r2_test)}")
```
